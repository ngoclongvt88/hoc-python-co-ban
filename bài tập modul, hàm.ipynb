```python
'''
Xây dựng hàm vẽ hình vuông với tham số là độ dài các cạnh. 
Sau khi hàm được xây dựng xong, tiến hành gọi hàm với các giá trị tham số khác nhau.
'''
def draw_square(n):
    import turtle
    wn = turtle.Screen()
    wn.bgcolor("blue")
    wn.title("star")
    
    pen = turtle.Turtle()
    pen.pensize(5)
    pen.color("white")
    pen.fillcolor("white")
    
    pen.right(45)
    pen.begin_fill()
    pen.circle(n, steps = 4)
    pen.end_fill()
    
    turtle.done()
n = int(input("nhập số n: "))
draw_square(100)
```

```python
'''
Xây dựng hàm vẽ hình đa giác đều với tham số là số cạnh của đa giác, và chiều dài của các cạnh. 
Sau khi hàm được xây dựng xong, tiến hành gọi hàm với các giá trị tham số khác nhau.
'''
def draw_polygon(number_of_edge, side_length):
    import turtle
    wn = turtle.Screen()
    wn.bgcolor("pink")
    wn.title("star")
    
    pen = turtle.Turtle()
    pen.pensize(5)
    pen.pencolor("blue")
    pen.fillcolor("blue")
    
    pen.right(45)
    pen.begin_fill()
    pen.circle(side_length, steps = number_of_edge)
    pen.end_fill()
    
    turtle.done()
draw_polygon(5, 100)    
```

```python
'''
Vẽ nhà từ các khối hình chữ nhật có kích thước và màu khác nhau. 
Hình dưới là một hình gợi ý, học viên được khuyến khích vẽ đẹp hơn.
'''
def draw_square1(length, width):
    import turtle
        #setup màn hình
    wn = turtle.Screen()
    wn.bgcolor("brown")
    wn.title("star")
        #setup chuột
    pen = turtle.Turtle()
    pen.pensize(5)
    pen.speed(100)
    pen.pencolor("white")
    pen.fillcolor("teal")
    
    #vẽ tầng1
    pen.penup()
    pen.goto(-300, -200)
    pen.pendown()
        
    pen.begin_fill()
    for i in range(2):
        pen.forward(length)
        pen.left(90)
        pen.forward(width)
        pen.left(90)    
    pen.end_fill()
    
    # cửa ra vào tầng 1
    pen.penup()
    pen.goto(-250, -200)
    pen.pendown()
 
    pen.fillcolor("#00FFFF")
    pen.begin_fill()
    for i in range(2):
        pen.forward(length/3)
        pen.pencolor("black")
        pen.left(90)
        pen.forward(width/2)
        pen.left(90)    
    pen.end_fill()
    
    pen.penup()
    pen.goto(-75, -200)
    pen.pendown()
    
    pen.pencolor("white")
    pen.begin_fill()
    for i in range(2):
        pen.forward(length/3)
        pen.pencolor("black")
        pen.left(90)
        pen.forward(width/2)
        pen.left(90)    
    pen.end_fill()
    
    #nhà để xe
    pen.penup()
    pen.goto(100, -200)
    pen.pendown()
    
    pen.fillcolor("#FF9900")
    pen.begin_fill()
    pen.pencolor("white")
    for i in range(2):
        pen.forward(length/2)
        pen.left(90)
        pen.forward(3*width/4)
        pen.left(90)    
    pen.end_fill()
    
    #cửa nhà để xe
    pen.penup()
    pen.goto(130, -200)
    pen.pendown()
    
    pen.fillcolor("#222222")
    pen.begin_fill()
    for i in range(2):
        pen.forward(length/3)
        pen.pencolor("blue")
        pen.left(90)
        pen.forward(width/3)
        pen.left(90)    
    pen.end_fill()
    
    #Tầng 2
    pen.penup()
    pen.goto(-300, -50)
    pen.pendown()
    
    pen.pencolor("white")
    pen.fillcolor("teal")
    pen.begin_fill()
    for i in range(2):
        pen.forward((2*length/3) + 50)
        pen.left(90)
        pen.forward(2*width/3)
        pen.left(90)    
    pen.end_fill()
    
    
    
    #tầng 3
    pen.penup()
    pen.goto(-300, 50)
    pen.pendown()
    
    pen.pencolor("white")
    pen.fillcolor("teal")
    pen.begin_fill()
    for i in range(2):
        pen.forward(length/2 + 50)
        pen.left(90)
        pen.forward(2*width/3)
        pen.left(90)    
    pen.end_fill()
    
    #vẽ cửa tầng2
    pen.penup()
    pen.goto(-270, -50)
    pen.pendown()
    
    pen.pencolor("white")
    pen.fillcolor("#FFFFCC")
    pen.begin_fill()
    for i in range(2):
        pen.forward(50)
        pen.left(90)
        pen.forward(70)
        pen.left(90)    
    pen.end_fill()
    
    pen.penup()
    pen.goto(-190, -20)
    pen.pendown()
    
    pen.right(45)
    pen.fillcolor("#66FF00")
    pen.begin_fill()
    pen.circle(30, steps = 4)
    pen.end_fill()
    
    pen.penup()
    pen.goto(-130, -20)
    pen.pendown()
    
    pen.fillcolor("#66FF00")
    pen.begin_fill()
    pen.circle(30, steps = 4)
    pen.end_fill()
    
    pen.penup()
    pen.goto(-60, -50)
    pen.pendown()
    
    pen.left(45)
    pen.pencolor("white")
    pen.fillcolor("#FFFFCC")
    pen.begin_fill()
    for i in range(2):
        pen.forward(50)
        pen.left(90)
        pen.forward(70)
        pen.left(90)    
    pen.end_fill()
    
    #vẽ cửa tầng 3
    pen.penup()
    pen.goto(-270, 50)
    pen.pendown()
    
    pen.pencolor("white")
    pen.fillcolor("#FFFFCC")
    pen.begin_fill()
    for i in range(2):
        pen.forward(50)
        pen.left(90)
        pen.forward(70)
        pen.left(90)    
    pen.end_fill()
    
    pen.penup()
    pen.goto(-180, 70)
    pen.pendown()
    
    pen.fillcolor("#66FF00")
    pen.begin_fill()
    for i in range(2):
        pen.forward(100)
        pen.left(90)
        pen.forward(50)
        pen.left(90)    
    pen.end_fill()
    
    
    turtle.done()    

draw_square1(400, 150) 
```
