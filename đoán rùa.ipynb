```python
'''
Tạo một số con rùa (giả sử 6 con rùa với các màu khác nhau) 
xuất phát tại vị trí bên trái của màn hình. 
Trước khi các con rùa chạy từ bên trái sang bên phải màn hình, 
cho phép người dùng đoán trước con rùa màu nào sẽ cán đích 
(cách vị trí xuất phát 1 khoảng cách nào đó) đầu tiên. 
Khi chạy, chương trình thay đổi vận tốc các chú rùa một cách 
ngẫu nhiên để kết quả con rùa cán đích sẽ là ngẫu nhiên. 
Chương trình kết thúc khi click trên màn hình.
'''
import turtle 
import random

#setup màn hình
wn = turtle.Screen()
wn.bgcolor("#00FFFF")
wn.setup(height = 500, width = 600)

#hiển thị cửa sổ nhập dự đoán của người chơi
player_guess = wn.textinput(prompt = "Hãy nhập màu rùa sẽ giành chiến thắng", title = "CHÚ Ý")
# Tạo list chưa màu các con rùa
list_color = ["blue", "black", "red", "pink", "yellow", "green", "purple"]
#tạo list chứa tốc độ các con rùa
list_speed = [0, 25, 30, 15 , 5, 20, 10]
#tạo vị trí xuất phát của Rùa trên trục Y
y_axis = [0, -75, 75, -150, 150, -225, 225, -300]
#tạo list rỗng để lưu các con rùa
list_turtle = []

#chương trình thiết lập vị trí ban đầu cho rùa và màu của các con rùa
for i in range(0, 7):
    pen = turtle.Turtle(shape = "turtle")
    pen.penup()
    pen.goto(-250, y_axis[i])
    pen.fillcolor(list_color[i])
    list_turtle.append(pen)
    
#tạo chương trình với vòng lặp vô hạn cho rùa chạy
run  = True

while run:
    for i in list_turtle:
        i.forward(random.choice(list_speed))
        if i.xcor() > 270:
            print(f"Rùa chiến thắng là rùa màu: {i}")
            run = False

#chương trình kết thúc khi click chuột lên màn hình
wn.exitonclick() 
```
