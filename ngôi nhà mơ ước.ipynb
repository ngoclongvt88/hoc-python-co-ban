```python
#vẽ ngôi nhà mơ ước
import turtle
pen = turtle.Turtle()

pen.pensize(5)
pen.pencolor("#808080")
pen.fillcolor("red")

# vex mai
mai_1 = 150
mai_2 = 300
pen.penup()
pen.goto(-300,100)
pen.pendown()
pen.begin_fill()
pen.left(45)
pen.forward(mai_1)
pen.right(45)
pen.forward(mai_2)
pen.right(135)
pen.forward(mai_1)
pen.right(45)
pen.forward(mai_2)
pen.end_fill()

#ve dau hoi
pen.pensize(10)
hoi = 100
pen.fillcolor("green")
pen.penup()
pen.goto(0,100)
pen.pendown()

pen.begin_fill()
pen.right(135)
pen.forward(mai_1)
pen.right(90)
pen.forward(mai_1)
pen.backward(10)

pen.pensize(5)
pen.right(45)
pen.forward(350)
pen.right(90)
pen.forward(195)
pen.right(90)
pen.forward(350)
pen.end_fill()

#than nha tang 2
pen.fillcolor("yellow")
pen.penup()
pen.goto(-290,100)
pen.pendown()
pen.backward(150)
pen.right(90)
pen.forward(300)

pen.penup()
pen.goto(-200,50)
pen.pendown
pen.right(45)
pen.begin_fill()
pen.circle(-50,steps=4)
pen.end_fill()

pen.penup()
pen.goto(-50,50)
pen.pendown
pen.begin_fill()
pen.circle(-50,steps=4)
pen.end_fill()

#ve mai tang 1
pen.fillcolor("red")
pen.penup()
pen.goto(-290,-50)
pen.pendown()
pen.begin_fill()
pen.left(45)
pen.forward(300)
pen.right(135)
pen.forward(50)
pen.right(45)
pen.forward(300)
pen.right(135)
pen.forward(50)
pen.end_fill()

pen.penup()
pen.goto(-25,-85)
pen.pendown()
pen.pensize(10)
pen.forward(180)
pen.right(90)
pen.forward(160)

#ve tang 1
pen.penup()
pen.goto(-315,-85)
pen.pendown()
pen.pensize(7)
pen.right(45)
pen.forward(155)
pen.left(90)
pen.forward(320)



turtle.done()
```
