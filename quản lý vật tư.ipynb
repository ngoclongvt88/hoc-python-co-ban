```python
from tkinter import *
from tkinter import ttk
import tkinter.messagebox
#import pymysql

class khi_tai:
    
    def __init__(seft, root):
        seft.root = root
        seft.root.geometry("1350x800+0+0")
        seft.root.title("Quản lý quân nhân")
        
        myframe = Frame(seft.root, bd=10, width=1350, height=700, relief=RIDGE, bg="cadet blue")
        myframe.grid()
#============================================Frame1===============================================================           
        frame1 = Frame(myframe, bd=10, width=1350, height=350, relief=RIDGE)
        frame1.grid(row=0, column=0)
        
        ho_ten_lb = Label(frame1, text="Họ và Tên", font=("arial", 14, "bold"))
        ho_ten_lb.grid(row=0, column=0)
        
        ho_ten_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        ho_ten_et.grid(row=0, column=1)
        
        nam_sinh_lb = Label(frame1, text="Năm sinh", font=("arial", 14, "bold"))
        nam_sinh_lb.grid(row=1, column=0)
        
        nam_sinh_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        nam_sinh_et.grid(row=1, column=1)
        
        cap_bac_lb = Label(frame1, text="Cấp bậc/Năm nhận", font=("arial", 12, "bold"))
        cap_bac_lb.grid(row=2, column=0)
        
        cap_bac_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        cap_bac_et.grid(row=2, column=1)
        
        vao_dang_lb = Label(frame1, text="Vào Đảng", font=("arial", 14, "bold"))
        vao_dang_lb.grid(row=3, column=0)
        
        vao_dang_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        vao_dang_et.grid(row=3, column=1)
        
        que_quan_lb = Label(frame1, text="Quê quán", font=("arial", 14, "bold"))
        que_quan_lb.grid(row=4, column=0)
        
        que_quan_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        que_quan_et.grid(row=4, column=1, columnspan=2)
        
        tru_quan_lb = Label(frame1, text="Nơi ở hiện tại", font=("arial", 14, "bold"))
        tru_quan_lb.grid(row=5, column=0)
        
        que_quan_et = Entry(frame1, font=("arial", 14, "bold"), bd="8")
        que_quan_et.grid(row=5, column=1, columnspan=2)
#============================================Frame2===============================================================       
        frame2 = Frame(myframe, bd=10, width=1350, height=300, relief=RIDGE)
        frame2.grid(row=2, column=0)
#============================================Frame3===============================================================           
        frame3 = Frame(myframe, bd=10, width=1350, height=50, relief=RIDGE)
        frame3.grid(row=1, column=0)
        
        
if __name__=="__main__":
    root = Tk()
    application = khi_tai(root)
    root.mainloop()
```
