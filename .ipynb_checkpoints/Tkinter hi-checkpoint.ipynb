```python
from tkinter import *

windows = Tk()
windows.title("Simple Calculator")
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

num1 = Entry(windows)
num2 = Entry(windows)

result_lbl = Label(windows, text='Result')

def add():
    result = int(num1.get()) + int(num2.get())
    result_lbl.config(text=result)
    
def subtract():
    result = int(num1.get()) - int(num2.get())
    result_lbl.config(text=result)
    
def multiply():
    result = int(num1.get()) * int(num2.get())
    result_lbl.config(text=result)

def divide():
    result = int(num1.get()) / int(num2.get())
    result_lbl.config(text=result)

add_btn = Button(windows, text='+', width=10, command=add)
minus_btn = Button(windows, text='-', width=10)
multiply_btn = Button(windows, text='*', width=10)
divide_btn = Button(windows, text='/', width=10)

num1.pack()
num2.pack()
add_btn.pack()
minus_btn.pack()
multiply_btn.pack()
divide_btn.pack()
result_lbl.pack()

windows.mainloop()
```

```python
from tkinter import *
from tkinter.ttk import *

windows = Tk()
windows.title("Registration")

registration_lbl = Label(windows, text="REGISTER")
fullname_lbl = Label(windows, text="Full name")
address_lbl = Label(windows, text="Address")
email_lbl = Label(windows, text="Email")

fullname_entry = Entry(windows)
address_entry = Entry(windows)
email_entry = Entry(windows)

btn_register = Button(windows, text="Register")
btn_login = Button(windows, text="Login")

combo = Combobox(windows, state="readonly")
combo['value'] = ("monday", "tuesday", "wednesday", "thursday")

text = Text(width=50, height=5)

registration_lbl.grid(row=0, column=0, columnspan=2)
fullname_lbl.grid(row=1, column=0)
email_lbl.grid(row=2, column=0)
address_lbl.grid(row=3, column=0)

fullname_entry.grid(row=1, column=1)
email_entry.grid(row=2, column=1)
address_entry.grid(row=3, column=1)

combo.grid(row=4, column=1)

text.grid(row=5, column=0, columnspan=3)

btn_login.grid(row=4, column=2)
btn_register.grid(row=4, column=3)


windows.mainloop()
```

```python
from tkinter import *

windows = Tk()
windows.title("Simple Calculator")
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

result_lbl = Label(windows, text="Result", width = 10)

num1 = Entry(windows)
num2 = Entry(windows)

def summation():
    result = int(num1.get()) + int(num2.get())
    result_lbl.config(text=result)
    
def subtraction():
    result = int(num1.get()) - int(num2.get())
    result_lbl.config(text=result)

def multiplication():
    result = int(num1.get()) * int(num2.get())
    result_lbl.config(text=result)
    
def division():
    result = int(num1.get()) / int(num2.get())
    result_lbl.config(text=result)
    
sum_btn = Button(windows, text="+", width=10, command=summation)
sub_btn = Button(windows, text = "-", width=10, command=subtraction)
mul_btn = Button(windows, text = "x", width=10, command=multiplication)
div_btn = Button(windows, text = "รท", width=10, command=division)

num1.pack()
num2.pack()
sum_btn.pack()
sub_btn.pack()
mul_btn.pack()
div_btn.pack()
result_lbl.pack()

windows.mainloop()
```
