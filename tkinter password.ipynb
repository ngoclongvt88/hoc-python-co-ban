```python
from tkinter import *
from tkinter import messagebox

windows = Tk()
windows.geometry("700x400")
windows.eval("tk::PlaceWindow . center")
windows.title("MyPass")

list_info =[]

def clear_value_etry():
    web_ety.delete(0, END)
    emai_use_ety.delete(0, END)
    pass_ety.delete(0, END)

def add_password():
    global list_info
    print(f"Website: {web_ety.get()}")
    print(f"Email/Username: {emai_use_ety.get()}")
    print(f"Password: {pass_ety.get()}")
    info = (web_ety.get(), emai_use_ety.get(), pass_ety.get())
    list_info.append(info)
    print(list_info)
    messagebox.showinfo("Add Password", "Password added successfully!!!!")
    clear_value_etry()
    
def search_info():
    print(f"website: {web_ety.get()}")
    for info in list_info:
        web_name =info[0]
        if web_name == web_ety.get():
            username_text = info[1]
            password_text = info[2]
            messagebox.showinfo("website info", f"Username: {username_text}, Password: {password_text}")
            break

mypass = Label(windows, text="MyPass")
mypass.grid(column=0, row=0, columnspan=2)
web_lbl = Label(windows, text="Website")
web_lbl.grid(column=0, row=1)
emai_use_lbl = Label(windows, text="Email/Username")
emai_use_lbl.grid(column=0, row=2)
pass_lbl = Label(windows, text="Password")
pass_lbl.grid(column=0, row=3)

web_ety = Entry(windows)
web_ety.grid(column=1, row=1)
emai_use_ety = Entry(windows)
emai_use_ety.grid(column=1, row=2)
pass_ety = Entry(windows)
pass_ety.grid(column=1, row=3)

add_but = Button(windows, text="Add", command=add_password)
add_but.grid(column=1, row=4, columnspan=3)
search_but = Button(windows, text="Search", command=search_info)
search_but.grid(column=4, row=1, columnspan=3)
gene_but = Button(windows, text="Generate Password")
gene_but.grid(column=4, row=3, columnspan=3)

windows.mainloop()
```
